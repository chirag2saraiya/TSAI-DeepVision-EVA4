Use split-folders package to prepare train and validation sets.politicians_dataset is already divided in train and test categories.


It must be noted here that in lfw dataset there are many classes for which single image is available. It was observed that when the dataset is split using 'split-folders' class having single image was being transferred to validation set and corresponding train class folder was kept empty. To take care of this situation folders are split in 'train = 0.3' and validation = '0.7' ratio and then the names are interchanged to get 70:30 train:val dataset split

```
import splitfolders
splitfolders.ratio("lfw_funneled", output="lfw_funneled_split", seed=1337, ratio=(.3, .7), group_prefix=None)
```

```
# exchange train and val folder names
!mv lfw_funneled_split/val lfw_funneled_split/train2
!mv lfw_funneled_split/train lfw_funneled_split/val
!mv lfw_funneled_split/train2 lfw_funneled_split/train
```
